```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={"Data":pd.to_datetime(["2022-01-01","2022-02-02","2022-03-02","2022-04-02","2022-05-02"]),
    "Value":[10,15,20,25,30]
}
df = pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='o')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with Trend")
plt.grid(True)
plt.show()
```
