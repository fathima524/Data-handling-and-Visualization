```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text_data="Data handling is the process of ensuring that research data is stored, archived or disposed off in a safe and secure manner during and after the conclusion of a research project"
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.show()
```

```python
words=text_data.split()
words_count=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(words_count.keys(),words_count.values())
plt.title('Historgram of word frequencies')
plt.xlabel('words')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show()
```
